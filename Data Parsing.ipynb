Цель - создание модели оценки справедливой стоимости квартир в Екатеринбурге.

## Шаг 2: сбор данных

Данные будем парсить с Циана, для этого воспользуемся уже готовой для этого библиотекой. Нас интересуют недорогие варианты, поэтому будем работать только с однокомнатными квартирами и студиями.

https://github.com/lenarsaitov/cianparser

```python
#pip install cianparser
```

```python
import cianparser
import pandas as pd
```

```python
data1 = cianparser.parse(
    deal_type="sale",
    accommodation_type="flat",
    location="Екатеринбург",
    rooms=1,
    start_page=1,
    end_page=50,
    is_saving_csv=False,
    is_latin=False,
    is_express_mode=False,
)
```

Библиотека имеет ограничение в 54 страницы сбора данных за раз, поэтому пройдемся по следующим 50 страницам, чтобы увеличить выборку.

```python
data2 = cianparser.parse(
    deal_type="sale",
    accommodation_type="flat",
    location="Екатеринбург",
    rooms=1,
    start_page=51,
    end_page=100,
    is_saving_csv=False,
    is_latin=False,
    is_express_mode=False,
)
```

И еще раз.

```python
data3 = cianparser.parse(
    deal_type="sale",
    accommodation_type="flat",
    location="Екатеринбург",
    rooms=1,
    start_page=101,
    end_page=150,
    is_saving_csv=False,
    is_latin=False,
    is_express_mode=False,
)
```

Преобразуем данные в датафреймы и объединим их.

```python
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df3 = pd.DataFrame(data3)

df = pd.concat([df1, df2, df3], axis=0)
df.head()
```

Сохраним данные в эксель и продолжим работу в другом блокноте.

```python
df.to_excel('output.xlsx', index=False)
```
